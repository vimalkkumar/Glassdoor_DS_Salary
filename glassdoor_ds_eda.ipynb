```python
cat_data = dataset[['Size', 'Type of ownership', 'Sector', 'Revenue', 'State', 'Same State', 'Python', 'R', 'AWS', 'Azure', 'Spark', 'Excel', 'Job Sample', 'Seniority']]
```

```python
for i in cat_data.columns:
    cat_num = cat_data[i].value_counts()
    print('Graph for %s: total = %d' % (i, len(cat_num)))
    fig = sbn.barplot(x = cat_num.index, y = cat_num)
    fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
    plt.show()
```

```python
for i in dataset[['Location' ,'Headquarters', 'Industry', 'Company Name Text']].columns:
    cat_num = dataset[i].value_counts()[:20]
    print('Graph for %s: total = %d' % (i, len(cat_num)))
    fig = sbn.barplot(x = cat_num.index, y = cat_num)
    fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
    plt.show()
```
